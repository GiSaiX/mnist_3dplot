```python
from pathlib import Path
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.manifold import TSNE
```

```python
def load_imgs(root, phase):
    imgs = []
    labels = []
    paths = root.glob(f"raw/{phase}/*/*.png")
    for path in paths:
        imgs.append(plt.imread(str(path)).reshape(1, 784))
        labels.append(np.array(int(path.parent.stem)).reshape(1,))
    return imgs, labels
```

```python
root = Path("/Users/takumi/datasets/MNIST")
imgs, labels = load_imgs(root, "test")
```

```python
imgs_reshaped = np.concatenate(imgs, axis=0)
print(imgs_reshaped.shape)
labels_reshaped = np.concatenate(labels, axis=0)
print(labels_reshaped.shape)

imgs_reduced = TSNE(n_components=3,  verbose=1).fit_transform(imgs_reshaped)
print(imgs_reduced.shape) 
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(imgs_reduced[:, 0], imgs_reduced[:, 1], imgs_reduced[:, 2], c=labels_reshaped)
fig.show()
```
